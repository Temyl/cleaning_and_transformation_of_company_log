```python
import pandas as pd
```

```python
# read the file
open_file = pd.read_csv('log.txt',delimiter='|')
open_file
```

```python
task_df = pd.DataFrame(open_file)
```

```python
# Rename the columns
task_df = task_df.rename(columns=
                {
    'Data quality assurance' : 'task',
    'success' : 'status',
    '89328463487' : 'tracker_ID',
    'Florida;USA' : 'server_region',
    '1599093305000' : 'date',
    '464f:bdbd:1d:535a:2dad:5117:a92f:b359' : 'server'
}
               )
```

```python
task_df
```

```python
#remove white spaces
def remove_white_spaces(text):
    if type(text) == str:
        return text.strip()
    return text

task_df['server_region'] = task_df['server_region'].apply(remove_white_spaces)

```

```python
# split thr server region to country and city
def city(text):
    results = text.split(';')[0]
    return results


task_df['city'] = task_df['server_region'].apply(city)

def country(text):
    results = text.split(';')[1]
    return results


task_df['country'] = task_df['server_region'].apply(country)

```

```python
# convert timestamp to date month,year format
task_df['date'] = pd.to_datetime(task_df['date']).dt.date
```

```python
# save to csv
columns = ['task', 'status', 'tracker_ID', 'date', 'server', 'city', 'country'] 
log_df = task_df[columns]

log_df.to_csv('log_of_details.csv')
```
